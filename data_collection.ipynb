```python
import akshare as ak
import pandas as pd
```

```python
funds_infos = ak.fund_name_em()
```

```python
etf_list = ['510210', '159903', '159915', '510300', '510500', '512100', '510880', '511220', '511020', '511030', '159972', '511380']
```

```python
data_list = {}
for etf in etf_list:
    etf_name = funds_infos.loc[funds_infos['基金代码'] == etf, '基金简称'].values[0]
    data = ak.fund_etf_hist_em(symbol=etf, period="daily", adjust="qfq")
    data_list[etf_name] = data
    start = data.loc[0, '日期']
    print(f'{etf_name}, {etf}, 成立时间, {start}')
```

```python
future_table = ak.futures_hist_table_em()
```

```python
future_table
```

```python
future_list = ['沪金主连', '螺纹钢主连', '沪铜主连', '豆粕主连', 'PTA主连', '十债主连']  # 贵金属，黑色，有色，农产品，能源化工
```

```python
for future in future_list:
    data = ak.futures_hist_em(symbol=future, period="daily")
    data_list[future] = data
    start = data.loc[0, '时间']
    print(f'{future}起始时间：{start}')
```

```python
data_list['沪金主连']
```

```python
assets = list(data_list.keys())
```

```python
assets
```

```python
with pd.ExcelWriter("data/assets V1.xlsx", engine="openpyxl", mode="a", if_sheet_exists="replace") as writer:
    for asset in assets:
        data = data_list[asset]
        data.to_excel(writer, index=False, sheet_name=asset)
```
